```python
from pydrive2.auth import GoogleAuth, ServiceAccountCredentials
from pydrive2.drive import GoogleDrive
import os

scope = ['https://www.googleapis.com/auth/drive']
gauth = GoogleAuth()
# gauth.settings['client_config_backend'] = 'service'
gauth.credentials = ServiceAccountCredentials.from_json_keyfile_name(
    'service_account.json', scope
)
drive = GoogleDrive(gauth)

folder_id = "1Z-tkJG0x-uCtOlPO28YX2mxe6YE8nFj3"
output_dir = "data"
os.makedirs(output_dir, exist_ok=True)

groups = ["Vibe Learning (profe agregué 3 escenarios)"]

def download_folder(folder_id, dest_path, allowed_folders=None):
    query = f"'{folder_id}' in parents and trashed=false"
    file_list = drive.ListFile({'q': query}).GetList()

    for f in file_list:
        name = f['title']
        local_path = os.path.join(dest_path, name)

        if f['mimeType'] == 'application/vnd.google-apps.folder':
            if allowed_folders is not None and name not in allowed_folders:
                continue

            print(f"[Carpeta] {local_path}")
            os.makedirs(local_path, exist_ok=True)

            download_folder(f['id'], local_path, allowed_folders=None)

        else:
            print(f"[Descargando] {local_path}")
            f.GetContentFile(local_path)

download_folder(folder_id, output_dir, allowed_folders=groups)
```
